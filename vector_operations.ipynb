```python
a = np.array([2, 3])
b = np.array([4, 4])
dot = np.dot(a, b)
print(dot)
```

```python
c = np.array([20,34,67])
d = np.array([30,44,54])
dot2 = np.dot(c, d)
print(dot2)
```

```python
a2 = np.array([2, 3])
b2 = np.array([4, 4])
cross = np.cross(a2, b2)
print(cross)
```

```python
c2 = np.array([20,34,67])
d2 = np.array([30,44,54])
cross2 = np.cross(c2, d2)
print(cross2)
```

```python
x = np.array([2, 5])
y = np.array([8, -6])
magy = np.sqrt(np.sum(y**2))
proj = (np.dot(x,y)/magy**2)*y
print(proj)
```

```python
x2 = np.array([2, 5, 10])
y2 = np.array([8, -6, 11])
magy = np.sqrt(np.sum(y2**2))
proj = (np.dot(x2,y2)/magy**2)*y2
print(proj)
```
